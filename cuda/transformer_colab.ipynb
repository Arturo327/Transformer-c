## 3. Subir archivos
Ejecuta la celda siguiente para subir tus archivos. cuda/main.cu, cuda/nn.cu, cuda/nn.h, data/processed/DIM256/token_books.bin, models/DIM256/pesos.bin


```
from google.colab import files
uploaded = files.upload()
import shutil, os
for fname in uploaded:
         if fname.endswith('.h'):
                 shutil.move(fname, f'include/{fname}')
         elif fname.endswith(('.c', '.cu')):
                 shutil.move(fname, f'cuda/{fname}')
         elif fname.endswith('.bin'):
                 shutil.move(fname, f'files/{fname}')
         else:
                 shutil.move(fname, f'exe/{fname}')
print('Archivos colocados:')
!find cuda include files -type f
```

```
# 4. Compilar todo
# nn.cu + main.c → exe/nn    (entrenamiento)
!nvcc -O3 -use_fast_math \
             -Iinclude \
             cuda/nn.cu cuda/main.cu \
             -lcublas -lm \
             -o exe/nn_cuda \
             2>&1

print('Compilación OK' if os.path.exists('exe/nn_cuda') else 'ERROR: exe/nn_cuda no existe')
```

```
# 8. Entrenar
# Colab corta sesiones largas.
# Para reanudar: pon el mismo número de epoch donde te quedaste.
!./exe/nn_cuda files/pesos.bin files/token_books.bin
files.download('files/pesos.bin')
```
